```python
'''Program to put some numbers into an array then create a new array of
their squares'''
import numpy
x = numpy.array([1.2,6.,15.,21.,3.1]) # Create an array from a list
y = numpy.zeros(len(x), dtype=float) # Create an array to hold the squares
for i in range(len(x)): # Loop over a range from 0 up to the length of x
    y[i] = x[i]**2 # Square each element of x and put result in y
print('Original array: ', x)
print('Squared array: ', y)
```

```python
'''Program to put some numbers into an array then create a new array of
their squares.'''
import numpy
x = numpy.array([1.2,6.,15.,21.,3.1]) # Create an array from a list
total = 0 # Initialise the total variable
for i in range(len(x)): # Loop over a range from 0 up to the length of x
    total = total + x[i] # Add the next value in the array into the total
print('The array: ', x)
print('The total: ', total)
```

```python
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# for 3D axis
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

```python
class Robot:
    def __init__(self, name, degree, address, campus):
        self.name = name
        self.degree = degree
        self.address = address
        self.campus = campus
        
    def introduce_self(self):
        print('My name is: ', self.name)  # this in java
```

```python
r1 = Robot("Rajendra", "Ph.D", "Stillwater", "OSU")
r2 = Robot("Urmila", "Masters", "Stillwater", "OSU")
```

```python
r1.introduce_self()
r2.introduce_self()
```
