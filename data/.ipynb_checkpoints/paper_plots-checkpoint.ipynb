```python
#!/usr/bin/env python

import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,
                               AutoMinorLocator)
%matplotlib inline
"""
#import csv
x = []
y = []

with open('ice_Ih.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    for row in plots:
        x.append(int(row[0]))
        y.append(int(row[1]))
"""
"""
import numpy as np
#x, y = np.loadcsv('ice_Ih.csv', delimiter=',', unpack=True)

import pandas as pd
%matplotlib inline

df = pd.read_csv("ice_Ih.csv")
df.head()
"""
"""
plt.plot(x,y, label='Loaded from file!')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Interesting Graph\nCheck it out')
plt.legend()
plt.show()
"""

import csv

with open('ice_XV.csv', 'r') as csvfile:
    readCSV = csv.reader(csvfile, delimiter=',')
    ring_size = []
    minimal_rings = []
    for row in readCSV:
        ring_size.append(int(row[0]))
        minimal_rings.append(float(row[1])/1000)

    # print(ring_size)
    # print(minimal_rings)
    
# Choose the position of each barplots on the x-axis (space=1,4,3,1)
# y_pos = [1, 3, 4, 5, 6, 7, 8, 9, 9]

# right font family
# plt.rc('font', family='serif')

fig = plt.figure(figsize=(4, 3))
ax = fig.add_subplot(1, 1, 1)

# create bars
ax.bar(ring_size[1:], minimal_rings[1:], width = 0.5, alpha = 1)

# to include font name, fontname = "Times New Roman"
plt.xlabel("Ring Size")
plt.ylabel(r'Average Number of Rings ( $\times$ 1000)')
plt.title("Minimal Ring Distribution of Ice XV\n1,250 Water Molecules")
plt.xlim(left = 2)   # to manage x-axis, plt.xlim(3, 10)
# plt.xlim(min(ring_size[1:]) * 1.1, max(ring_size[1:] * 1.1)
plt.xticks(np.arange(min(ring_size[:]), max(ring_size[1:]) + 1, 1.0))  # --> works in x-axis

plt.ylim(0, 4)

# to put minor axis
# import matplotlib.ticker as tck
# ax.yaxis.set_minor_locator(tck.AutoMinorLocator())

# Tick Parameters
plt.tick_params(axis='both', left='on', top='on', right='on',
                    bottom='on', labelleft='on', labeltop='off',
                    labelright='off', labelbottom='on', direction = 'in')


# Make a plot with major ticks that are multiples of 20 and minor ticks that
# are multiples of 5.  Label major ticks with '%d' formatting but don't label
# minor ticks.
ax.yaxis.set_major_locator(MultipleLocator(0.5))
ax.yaxis.set_major_formatter(FormatStrFormatter('%0.1f'))

# For the minor ticks, use no labels; default NullFormatter.
ax.yaxis.set_minor_locator(MultipleLocator(0.1))

# saving figure
plt.savefig("ice_XI.png", bbox_inches="tight", dpi = 300)  # if colorful background --> transparent = True

# high quality pdf file
# plt.savefig("ice.pdf")

# for latex image file
# plt.savefig('destination_path.eps', format='eps')

plt.show()
```
