```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
# plt.style.use('ggplot')
plt.style.use('seaborn-deep')

# set font
#plt.rcParams['font.family'] = 'sans-serif'
# plt.rcParams['font.sans-serif'] = 'Helvetica'

# set the style of the axes and the text color
plt.rcParams['axes.edgecolor']='#333F4B'
plt.rcParams['axes.linewidth']=0.8
plt.rcParams['xtick.color']='#333F4B'
plt.rcParams['ytick.color']='#333F4B'
plt.rcParams['text.color']='#333F4B'

company = ['Ice Ih', 'Ice II', 'Ice III', 'Ice IV']
ypos = np.arange(len(company))
ypos
revenue = [90, 136, 89, 27]
std_error = (5, 8, 12, 6)

# For names
# 1. Adjust x axis Ticks

# Create names on the x-axis
plt.xticks(ypos, company, fontsize = 8, rotation = 30, ha = 'center', va = 'top')

plt.ylabel("revenue(bin)", fontweight='black', color = '#333F4B')
plt.title("Ice formation", fontsize = 12)

# create bars
plt.bar(ypos, revenue, color = 'tomato', align = 'center', alpha = 1, label = "Revenue", yerr = std_error)

# fig = plt.bar(ypos, revenue, color = 'tomato', align = 'center', alpha = 1, label = "Revenue", yerr = std_error)
# fig.show()


# 2. Tick Parameters
# plt.tick_params(axis = 'both', which = 'major', labeltop = False, labelright = True, labelsize = 10) # size of label
# plt.tick_params(axis='both',bottom=True, top=True, left=True, right=True, direction='in', which='major', grid_color='blue')
plt.tick_params(axis='both', left='on', top='off', right='on',
                    bottom='on', labelleft='on', labeltop='off',
                    labelright='on', labelbottom='on')

# rotate axis labels
plt.setp(plt.gca().get_xticklabels(), rotation=45, horizontalalignment='right')

# Modify legend
plt.legend(   # plot items
           ['sine curve'],  
           frameon=True,                                   # legend border
           framealpha=1,                                   # transparency of border
           ncol=1,                                         # num columns
           shadow=True,                                    # shadow on
           borderpad=1,                                    # thickness of border
           title='Sines and Cosines')                      # title

plt.grid(linestyle='--', linewidth= 0.5, alpha=0.15)

# show graphics
plt.show()

# saving figure
plt.savefig("output.png", bbox_inches="tight") 
```

```python
import pandas as pd

df = pd.DataFrame({'contribution': [0.188137,0.160208,0.160208,0.151654,0.149489,0.135975,0.063206]}
                  ,index=['A','B','C','D','E','F','G'])

colors = ['b', 'green', 'y', 'pink','orange','cyan','darkgrey']

ax = df.plot(kind='barh', color=colors, legend=False)

for bar in ax.patches:
    bar.set_facecolor('#888888')

highlight = 'D'
pos = df.index.get_loc(highlight)

ax.patches[pos].set_facecolor('#aa3333')
ax.legend()
plt.savefig("output.png", bbox_inches="tight") 
```
